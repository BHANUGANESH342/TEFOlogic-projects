```python
y=fi['final_grade']
x=fi.drop(['final_grade'],axis=1)
```

```python
x
```

```python
y
```

```python
#spliting data to create the model
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_test.describe()
```

```python
x_train.describe()
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
ypred=reg.predict(x_test)
ypred
```

```python
from sklearn.metrics import r2_score #to check the efficiency
r2_score(y_test,ypred)
```

```python
from sklearn.metrics import mean_squared_error #calculate MSE
mean_squared_error(ypred,y_test)
```

```python
results=r.DataFrame(columns=['final_grade','Predicted']) #To compare the actual and predicted price
results['final_grade']=y_test
results['Predicted']=ypred
results
```

```python
results['Difference']=results.apply(lambda row:row.Predicted - row.final_grade,axis=1)
results
```
